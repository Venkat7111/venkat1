```python
df.rename(columns={'2urvived': 'Survived'}, inplace=True)

# Step 2: Calculate IQR
Q1 = df['Age'].quantile(0.25)
Q3 = df['Age'].quantile(0.75)
IQR = Q3 - Q1

# Step 3: Define acceptable range
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 4: Filter the data to remove outliers
df_no_outliers = df[(df['Age'] >= lower_bound) & (df['Age'] <= upper_bound)]

# Step 5: Boxplot without outliers
sns.boxplot(data=df_no_outliers, x="Survived", y="Age")
plt.title("Age Distribution by Survival (Outliers Removed)")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```
